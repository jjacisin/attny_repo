```python
#return yankeees managers
import requests
from bs4 import BeautifulSoup
import re ##IMPORTS REGEX "regluar expression";


class BBScraper:
    def webpage_html(self, url = 'https://www.baseball-reference.com/teams/NYY/'):
        person_request = requests.get(url)
        self.person_html = person_request.text
        return self.person_html

    def listings_html(self, person_html = None):
        team_html = person_html or self.person_html #takes in a new html code or uses orginsl HTML
        team_soup = BeautifulSoup(person_html) #creates BeautifulSoup object
        team_stats_soup =  person_soup.findAll('table', {'class':"sortable"})
#         self.attorney_soup = attorney_soup
        return team_stats_soup
```
